```python
from nltk.tokenize import word_tokenize
import csv

EVpath = '/content/drive/MyDrive/Colab Notebooks/ClassEval.csv'

EVarray = []

# csv읽고 전처리
with open(EVpath, 'r') as f1:
    file_read = csv.reader(f1)
    for line in f1:
        EVarray.append(line)
    
    evalList = [[''] for _ in range(len(EVarray))]
    for i in range (len(EVarray)):
        evalList[i] += EVarray[i].split(',')

SWpath = '/content/drive/MyDrive/Colab Notebooks/stopwords.csv'

SWarray = []

# csv읽고 전처리
with open(SWpath, 'r') as f2:
    file_read = csv.reader(f2)
    for line in f2:
        SWarray.append(line)
    
    stopWords = []
    for i in range (1,len(SWarray)): 
        stopWords += SWarray[i].split(',')
    
    for i in range(1, len(stopWords)):
        stopWords[i] = stopWords[i].replace("\n", "") 
```

```python
yong = [[''] for _ in range(len(evalList))]
for i in range(1, len(evalList)):
  yong[i] = evalList[i][4:]
```

```python
yong2 = [[''] for _ in range(len(yong))]
for i in range(1, len(yong)):
  yong2[i] = ','.join(yong[i])
for i in range(1, len(yong2)):
  yong2[i] = yong2[i].replace("\n", "") 
  yong2[i] = yong2[i].replace('"', "")
  yong2[i] = yong2[i].replace('..', "")
  yong2[i] = yong2[i].replace(',,', "")
  yong2[i] = yong2[i].replace(',', "")
  yong2[i] = yong2[i].replace('.', "") 
```

```python
del yong2[0]
```

```python
#전처리 함수
!pip install git+https://github.com/haven-jeon/PyKoSpacing.git

from konlpy.tag import Kkma
import sys
import os
import re
sys.path.append(os.path.dirname('PyKoSpacing/'))
# from pykospacing import spacing

kkma = Kkma()

def preprocessing(review):
  i=0
  total_review = ''
  
  for s in kkma.sentences(review):
    if i == 7:
      break
    
    s = re.sub('([a-zA-Z])','',s)
    s = re.sub('[ㄱ-ㅎㅏ-ㅣ]+','',s)
    s = re.sub('[-=+,#/\?:^$@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','',s)
    
    # if len(s) < 198:
    #   s = spacing(s)
    s += '. '
    total_review += s
  
    i+=1
  return total_review
```

```python
from gensim.summarization.summarizer import summarize
sumEval = [[''] for _ in range(len(yong2))]

for i in range(700):
  if len(result[i].split('.')) < 4:
    sum = re.sub('"', '', evalList[i+1][4].split('.')[0])
    
  else:
    
    sum = summarize(result[i], word_count = 15)
    sum = re.sub('\n', ' ', sum)

  if sum == '':
    sum = re.sub('"', '', evalList[i+1][4].split('.')[0])
  
  sumEval[i] = (sum);
```
